Dari dataset yang diambil, diputuskan untuk mengambil bebrapa kolom yang bisa dijadikan kriteria untuk sistem pendukung keputusan rekomendasi perusahaan terbaik untuk investor
1. Rating: Memberikan gambaran tentang kualitas kredit dan risiko perusahaan menurut lembaga pemeringkat independen.

2. Debt/Equity Ratio: Menunjukkan struktur modal perusahaan dan tingkat leverage keuangan. Rasio ini penting untuk menilai risiko keuangan perusahaan.

3. Net Profit Margin: Mengukur efisiensi perusahaan dalam menghasilkan laba bersih dari pendapatannya. Ini menunjukkan profitabilitas keseluruhan perusahaan.

4. ROE - Return On Equity: Mengukur seberapa efisien perusahaan menggunakan ekuitas pemegang saham untuk menghasilkan laba. Ini adalah indikator penting bagi investor untuk menilai kinerja manajemen dalam menggunakan modal.

Keempat kriteria ini memberikan gambaran yang cukup komprehensif tentang kesehatan keuangan, risiko, profitabilitas, dan efisiensi manajemen perusahaan, yang sangat relevan bagi investor dalam membuat keputusan investasi.

Dan dari keempat kriteria tersebut telah ditentukan cost dan benefitnya
1. ROE - Return On Equity
   - Benefit: Mengukur efisiensi perusahaan dalam menghasilkan keuntungan dari modal pemegang saham.
   - Nilai yang lebih tinggi menunjukkan perusahaan lebih efisien dalam menggunakan ekuitas untuk menghasilkan laba.

2. Net Profit Margin
   - Benefit: Menunjukkan persentase pendapatan yang tersisa sebagai laba bersih setelah semua biaya dan pajak.
   - Margin yang lebih tinggi menandakan profitabilitas yang lebih baik.

3. Operating Cash Flow Per Share
   - Benefit: Menggambarkan kemampuan perusahaan menghasilkan kas dari operasi bisnisnya per saham.
   - Nilai yang lebih tinggi menunjukkan kinerja operasional yang lebih kuat.

Cost (kriteria yang lebih rendah nilainya lebih baik):

1. Debt/Equity Ratio
   - Cost: Mengukur tingkat leverage keuangan perusahaan.
   - Rasio yang lebih rendah menunjukkan risiko keuangan yang lebih kecil dan struktur modal yang lebih konservatif.

untuk bobot dari setiap kriteria
1. ROE - Return On Equity (Benefit)
   Bobot: 30%
   Alasan: ROE adalah indikator kunci efisiensi perusahaan dalam menghasilkan laba dari modal pemegang saham. Ini sangat penting bagi investor karena langsung menunjukkan tingkat pengembalian investasi mereka.

2. Net Profit Margin (Benefit)
   Bobot: 25%
   Alasan: Net Profit Margin menunjukkan efisiensi operasional dan kemampuan perusahaan menghasilkan laba dari pendapatannya. Ini penting untuk menilai profitabilitas keseluruhan, namun sedikit di bawah ROE dalam hal kepentingan langsung bagi investor.

3. Operating Cash Flow Per Share (Benefit)
   Bobot: 20%
   Alasan: Arus kas operasional adalah indikator penting kesehatan finansial dan kemampuan perusahaan menghasilkan kas dari operasi utamanya. Meski penting, ini diberi bobot lebih rendah karena bisa berfluktuasi lebih banyak dalam jangka pendek.

4. Debt/Equity Ratio (Cost)
   Bobot: 25%
   Alasan: Rasio ini sangat penting untuk menilai risiko finansial perusahaan. Struktur modal yang sehat adalah kunci keberlanjutan jangka panjang, sehingga diberi bobot yang cukup signifikan.

Total bobot: 100%


- Bobot yang lebih tinggi diberikan pada ROE dan Net Profit Margin karena keduanya langsung berkaitan dengan profitabilitas, yang umumnya menjadi fokus utama investor.
- Operating Cash Flow Per Share mendapat bobot sedikit lebih rendah karena, meski penting, arus kas bisa lebih berfluktuasi dan tidak selalu mencerminkan profitabilitas jangka panjang.
- Debt/Equity Ratio mendapat bobot yang cukup signifikan karena risiko finansial sangat penting dalam menilai kesehatan dan keberlanjutan perusahaan.

Data Preparation

Setelah menentukan kriteria dan bobot dari dataset selanjutnya akan dilakukan tahap data preparation

karna kolom kriteria sudah ditentukan, maka kolom yang tidak termasuk kriteria akan didrop/hapus

```python
df = df.loc[:,['Corporation','ROE - Return On Equity','Net Profit Margin','Operating Cash Flow Per Share','Debt/Equity Ratio']];

df.head()
```

meilihat adakah data yang kosong pada dataframe 

```python
print(df.isna().sum())
```

Karna data yang diambil merupakan gabungan data dari bebrapa lembaga survei jadi ada data yang memiliki nilai yang sama, data ini yang harus dibersihkan juga

```python
duplicates_specific = df[df.duplicated(subset=['Corporation'], keep=False)]
print("\nJumlah baris duplikat:", duplicates_specific.shape[0])
```

Setelah ditemukan data yang duplicate, bisa dilakuman pembersihan data

```python
df_clean = df.drop_duplicates(subset=['Corporation'])
print(df_clean)
```

Selanjutnya data yang sudah dibersihkan akan dimasukkan ke file csv untuk selanjutnya di proses

```python
df_clean.to_csv('dataset/dataset_clean.csv', index=False)
```
